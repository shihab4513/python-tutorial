```python
#Get the characters from the start to position 5 (not included)
b = "Hello22 World!"
print(b[:5])
```

```python
#here -5 means index 5 from right side . so here -5 index is W and -2 index is l . so its excluding -5 part but including -2 part
b = "Hello, World!"
print(b[-5:-2])
```
